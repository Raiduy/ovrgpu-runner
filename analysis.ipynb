```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('data/2023-11-29T01.40.16.5122571+01.00.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df
```

```python
df.columns.to_list()
```

```python
# plot timestamp against all columns in separate plots
for col in df.columns.to_list():
    if col != 'timestamp':
        sns.lineplot(x='timestamp', y=col, data=df)
        plt.show()
        
```

```python
# correlation matrix
corr = df.corr()
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(corr)

# save figure as png
plt.savefig('correlation_matrix.png')
```
